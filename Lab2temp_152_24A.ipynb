```python
#start code here
x, y , = sp.symbols('x y ', real = True, positive = True)

formula = 2*sp.pi*(y*(2*sp.sqrt(1-y**2)))

display(sp.integrate(formula, (y, 0, 1)))
```

#### 1b

```python
#start code here
r = sp.Symbol('r', real = True)
formula = 2*sp.pi*(y*(2*sp.sqrt(1-y**2)))
int_formula = sp.integrate(formula, (y, 0, r))
result = sp.nsolve(int_formula - 80,r,1)
print('The radius is:')
display(sp.re(result))
```

```python
#start code here
x = sp.symbols('x', real = True)
f_x = 2*x**2
g_x = abs(x**2 - 9)

p1 = plot(f_x, (x, -5, 5), show=False, line_color='r', legend=True, label='f(x) = 2x^2')
p2 = plot(g_x, (x, -5, 5), show=False, line_color='b', legend=True, label='g(x) = |x^2 - 9|')
p1.extend(p2)
p1.show()
```
