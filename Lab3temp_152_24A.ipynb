```python
# Enter your Python code here
#washer method:
bounds = sp.solve(g_x - f_x)
ac = [u.evalf() for u in bounds]
washer = sp.integrate(2*sp.pi*((x-1)- (f_x-g_x)), (x,ac[1],ac[0]))
print('The value of the region by the two curves about x = 1 is: ')
display(washer)
```
